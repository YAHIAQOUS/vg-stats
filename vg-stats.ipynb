## 1. Which company is the most common video game publisher?

```python
common_publisher=df['Publisher'].value_counts()
common_publisher.head(1)
```

## 2. What’s the most common platform?

```python
common_platform=df['Platform'].value_counts()
common_platform.head(1)
```

## 3. What about the most common genre?

```python
common_genre=df['Genre'].value_counts()
common_genre.head(1)
```

## 4. What are the top 20 highest grossing games?

```python
highest_sales = df[['Name','Global_Sales']].sort_values('Global_Sales',ascending=False)

highest_sales.head(20)
```

## 5. For North American video game sales, what’s the median?

```python
median_north_american = df[['Name','NA_Sales']].sort_values('NA_Sales',ascending=False).median()
median_north_american
```

```python
median_north_american = df[['Name','NA_Sales']].sort_values('NA_Sales',ascending=True).median()
median_north_american
```

## 6. For the top-selling game of all time, how many standard deviations above/below the mean are its sales for North America?

## 7. The Nintendo Wii seems to have outdone itself with games. How does its average number of sales compare with all of the other platforms?

```python
average_publisher_sales= df[['Publisher','Global_Sales']].mean()
average_publisher_sales
```

```python
max_publisher_sales= df[['Publisher','Global_Sales']].groupby('Publisher').mean().sort_values('Global_Sales',ascending=False).head()
max_publisher_sales
```
